```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
probe_training_rate = np.array([200e-1, 100e-1, 70e-1, 50e-1, 30e-1, 20e-1, 10e-1, 1e-1, 1e-2])
loss_after_20_iterations = np.array([1.10e5, 2.696e4, 1.95e4,1.52e4, 1.50e4, 3.567e4, 1.01e5, 4.49e5, 5.73e5])

plt.plot(probe_training_rate, loss_after_20_iterations, marker='o')
plt.xlabel('Probe Training Rate')
plt.ylabel('Loss after 20 iterations')
plt.xscale('log')
plt.yscale('log')
plt.plot()
```

```python
theta_phi_training_rate = np.array([25e-1, 15e-1, 8e-1, 5e-1, 2e-1, 1e-1, 5e-2, 1e-2])
loss_after_20_iterations = np.array([1.15e2, 5.756e1, 1.699e1, 1.41e1, 1.409e1, 1.55e1, 3.779e1, 2.25e2])

plt.plot(theta_phi_training_rate, loss_after_20_iterations, marker='o')
plt.xlabel('Theta-Phi Training Rate')
plt.ylabel('Loss after 20 iterations')
plt.xscale('log')
plt.yscale('log')
plt.plot()
```
