```python
def normalise_neel(l):
    # Normalize to ensure L is always a unit vector
    mag = torch.sqrt(torch.sum(l**2, dim=0) + 1e-8)
    Lx = l[0] / mag
    Ly = l[1] / mag
    Lz = l[2] / mag
    return Lx, Ly, Lz

def optimize_object(scan, pol_angles, I_meas, H, W, initial_l, initial_probe_amplitude,num_iterations=100, device=None,
                                             optimizer_params={'l_lr': 1e-1, 'CA_lr': 1e-5},
                                             plot_filename=None):
    """
    Reconstruct probes (different for each polarisation) l C, A1, A2 from measured diffraction patterns.
    This is different to previous versions because it directly updates lx, ly, lz to avoid differentiation problems associated with angular coordinates.
    .
    Parameters
    ----------
    scan : ScanTrajectory
        Scan positions.
    pol_angles : list of float
        List of polarisation angles in degrees for each probe.
    I_meas : torch.Tensor
        Measured diffraction intensities, shape (N_probes, N_scan, Hdet, Wdet)
    H, W : int
        Object map size.
    initial_l : torch.Tensor
        Initial guess for the neel vector in Cartesian coordinates, shape (3, H, W).
    initial_probe_amplitude : torch.Tensor
        Initial guess for the probe amplitude, shape (H, W).
    num_iterations : int
        Number of optimization steps.
    device : torch.device or str
        Device to use.
    loss_prefactors : dict
        Prefactors for different loss components - how much to weight anisotropy and gradient losses relative to the amplitude loss.
    optimizer_params : dict
        Learning rates for different parameter groups in the optimizer.
    plot_filename : str or None
        If not None, save intermediate plots to this filename during optimization.
     Returns
    -------
    dict
        Reconstructed parameters and loss history.
    """
    if device is None:
        device = I_meas.device

    # This is for the live plot updates
    plot_update = create_live_plotter(Lx, Ly, phi_cmap=RGB_scale, theta_cmap='magma')

    eps = 1e-12
    cdtype = torch.complex64

    # Learnable parameters

    l = torch.nn.Parameter(initial_l) # Defining the neel vector in terms of it's lx, ly, lz components, which are then normalised to get theta and phi. This is easier to optimise than directly optimising theta and phi, which have singularities at the poles.
    probe_amplitude = initial_probe_amplitude

    # Initialise the probe array (same for all pol states at the moment)
    probes = []
    for angle in pol_angles:
        rad = np.deg2rad(angle)
        jones_vec = torch.tensor([np.cos(rad) + 0j, np.sin(rad) + 0j], dtype=cdtype, device=device)
        probes.append(Probe(probe_amplitude, jones_vec, fluence=fluence, normalized=False))
    probe_amplitude = (probes[0].amplitude) #It has now been normalised to the correct fluence.
    probe_amplitude_start = probes[0].amplitude
    fluence_calc = torch.sum(torch.abs(probe_amplitude_start)**2)
    print(f"Calculated initial fluence from probe amplitude: {fluence_calc.item():.6e}, target fluence: {fluence.item():.6e}")

    optimizer = torch.optim.AdamW([
        {"params": [l], "lr": optimizer_params['l_lr']},
        {"params": [C, A1, A2], "lr": optimizer_params['CA_lr']}
        ])
    
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer,
        mode='min',
        factor=0.5,       # reduce LR by 2x
        patience=5,      # wait 5 iterations without improvement
        threshold=1e-3,   # minimum change to count as improvement
        min_lr=1e-6,      # don't go too small
    )
    loss_history = []

    for iteration in range(num_iterations):
        optimizer.zero_grad(set_to_none=True)

        l_mag = torch.sqrt(torch.sum(l**2, dim=0) + 1e-8) 
        
        # Create normalized Cartesian components
        lx_norm = l[0] / l_mag
        ly_norm = l[1] / l_mag
        lz_norm = l[2] / l_mag

        # Build Jones object from current parameters
        neel = NeelObject(C, A1, A2)
        J = neel.build_jones_from_cartesian(lx= lx_norm, ly=ly_norm, lz=lz_norm)
        obj = JonesObject(J)

        # Remake the probe array
        probes = []
        for angle in pol_angles:
            rad = np.deg2rad(angle)
            jones_vec = torch.tensor([np.cos(rad) + 0j, np.sin(rad) + 0j], dtype=cdtype, device=device)
            probes.append(Probe(probe_amplitude, jones_vec, fluence=fluence, normalized=True))

        # Forward model
        model = ForwardModel(obj, Propagator(), Detector())
        I_pred = model.simulate_all(probes, scan)
        eps = 1e-8

        fluence_calc = torch.sum(torch.abs(probe_amplitude)**2)
        print(f"Calculated fluence from probe amplitude: {fluence_calc.item():.6e}, target fluence: {fluence.item():.6e}")

        # Define the loss function
        loss = torch.mean((torch.sqrt(I_pred + eps) - torch.sqrt(I_meas + eps))**2)

        loss.backward()
        optimizer.step()

        # Step the scheduler with the current loss
        scheduler.step(loss)

        if iteration % 20 == 0:
            print([group['lr'] for group in optimizer.param_groups])

        loss_history.append(loss.item())

        if iteration % 1 == 0:
            theta = torch.acos(lz_norm)
            phi = torch.atan2(ly_norm, lx_norm)
            l_mag = torch.sqrt(torch.sum(l**2, dim=0) + 1e-8)
            print(f"Iter {iteration:4d} | Loss = {loss.item():.6e}")
            plot_update(probe_amplitude, theta, phi, loss, save_filename=plot_filename)

    return {
        "l": l.detach(),
        "probe_amplitude": probe_amplitude.detach(),
        "C": C.detach(),
        "A1": A1.detach(),
        "A2": A2.detach(),
        "loss_history": loss_history,
        "optimizer_params": optimizer_params  
    }

'''
initial_l = torch.stack([
    torch.randn(H, W, device=device)*0.1,
    torch.randn(H, W, device=device)*1,
    torch.randn(H, W, device=device)*0.1
])
'''
initial_l = torch.randn(3, H, W, device=device)
result = optimize_object(scan, pol_angles, I_sim, H, W,
                                initial_probe_amplitude=P,
                                initial_l=initial_l,
                                num_iterations=5, device=device)
```

```python
def optimize_object_and_probe_with_scheduler(scan, pol_angles, I_meas, H, W, initial_l, initial_probe_amplitude,num_iterations=100, device=None, 
                                             loss_prefactors={'sqrt_amp_pf': 1.0, 'gradient_pf': 1e-4, 'anisotropy_pf': 1e-4, 'probe_localisation_pf': 1e-3},
                                             optimizer_params={'l_lr': 1e-3, 'CA_lr': 1e-5, 'probe_lr': 60e-1},
                                             plot_filename=None, object_randomisation=False, probe_randomisation=False):
    """
    Reconstruct probes (different for each polarisation) l C, A1, A2 from measured diffraction patterns.
    This is different to previous versions because it directly updates lx, ly, lz to avoid differentiation problems associated with angular coordinates.
    .
    Parameters
    ----------
    scan : ScanTrajectory
        Scan positions.
    pol_angles : list of float
        List of polarisation angles in degrees for each probe.
    I_meas : torch.Tensor
        Measured diffraction intensities, shape (N_probes, N_scan, Hdet, Wdet)
    H, W : int
        Object map size.
    initial_l : torch.Tensor
        Initial guess for the neel vector in Cartesian coordinates, shape (3, H, W).
    initial_probe_amplitude : torch.Tensor
        Initial guess for the probe amplitude, shape (H, W).
    num_iterations : int
        Number of optimization steps.
    device : torch.device or str
        Device to use.
    loss_prefactors : dict
        Prefactors for different loss components - how much to weight anisotropy and gradient losses relative to the amplitude loss.
    optimizer_params : dict
        Learning rates for different parameter groups in the optimizer.
    plot_filename : str or None
        If not None, save intermediate plots to this filename during optimization.
    object_randomisation : bool
        Whether to apply randomisation to the object parameters during optimization to help escape local minima.
    probe_randomisation : bool
        Whether to apply randomisation to the probe parameters during optimization to help escape local minima.
     Returns
    -------
    dict
        Reconstructed parameters and loss history.
    """
    if device is None:
        device = I_meas.device

    # This is for the live plot updates
    plot_update = create_live_plotter(Lx, Ly, phi_cmap=RGB_scale, theta_cmap='magma')

    eps = 1e-6
    cdtype = torch.complex64

    # Learnable parameters

    l = torch.nn.Parameter(initial_l) # Defining the neel vector in terms of it's lx, ly, lz components, which are then normalised to get theta and phi. This is easier to optimise than directly optimising theta and phi, which have singularities at the poles.
    probe_amplitude = initial_probe_amplitude

    # Initialise the probe array (same for all pol states at the moment)
    probes = []
    for angle in pol_angles:
        rad = np.deg2rad(angle)
        jones_vec = torch.tensor([np.cos(rad) + 0j, np.sin(rad) + 0j], dtype=cdtype, device=device)
        probes.append(Probe(probe_amplitude, jones_vec, fluence=fluence, normalized=False))
    probe_amplitude = torch.nn.Parameter(probes[0].amplitude) #It has now been normalised to the correct fluence.
    probe_amplitude_start = probes[0].amplitude
    fluence_calc = torch.sum(torch.abs(probe_amplitude_start)**2)
    print(f"Calculated initial fluence from probe amplitude: {fluence_calc.item():.6e}, target fluence: {fluence.item():.6e}")

    optimizer = torch.optim.AdamW([
        {"params": [l], "lr": optimizer_params['l_lr']}, #{"params": [C, A1, A2], "lr": optimizer_params['CA_lr']},
        {"params": [probe_amplitude], "lr": optimizer_params['probe_lr']}
        ])
    
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer,
        mode='min',
        factor=0.5,       # reduce LR by 2x
        patience=500,      # wait 5 iterations without improvement
        threshold=1e-3,   # minimum change to count as improvement
        min_lr=1e-6,      # don't go too small
    )
    loss_history = []

    for iteration in range(num_iterations):
        STXM_end_iteration = 50
        if iteration < STXM_end_iteration:
            for p in [l, C, A1, A2]:
                p.requires_grad_(True)
            probe_amplitude.requires_grad_(False)
            loss_prefactors['STXM_pf_c'] = loss_prefactors['STXM_pf']#/((iteration+1)/STXM_end_iteration)
            loss_prefactors['sqrt_amp_pf_c'] = loss_prefactors['sqrt_amp_pf']#*(iteration+1)/STXM_end_iteration
            print('Free object parameters, fixing probe amplitude.')
        '''
        if iteration >= STXM_end_iteration and iteration < 70:
            for p in [l, C, A1, A2]:
                p.requires_grad_(False)
            probe_amplitude.requires_grad_(True)
            loss_prefactors['sqrt_amp_pf_c'] = 1.0
            loss_prefactors['STXM_pf_c'] = 0.0
            print('Free probe, fixed object')
        '''
        if iteration >= 50 and iteration % 2 == 0:
            for p in [l, C, A1, A2]:
                p.requires_grad_(True)
            probe_amplitude.requires_grad_(False)
        if iteration >= 50 and iteration % 2 == 1:
            for p in [l, C, A1, A2]:
                p.requires_grad_(False)
            probe_amplitude.requires_grad_(True)
        if iteration >= 80:
            for p in [l, C, A1, A2]:
                p.requires_grad_(True)
            probe_amplitude.requires_grad_(True)
            print('Freeing all parameters.')
        
        optimizer.zero_grad(set_to_none=True)

        l_mag = torch.sqrt(torch.sum(l**2, dim=0) + 1e-8) 
        
        # Create normalized Cartesian components
        lx_norm = l[0] / l_mag
        ly_norm = l[1] / l_mag
        lz_norm = l[2] / l_mag

        # Build Jones object from current parameters
        neel = NeelObject(C, A1, A2)
        J = neel.build_jones_from_cartesian(lx= lx_norm, ly=ly_norm, lz=lz_norm)
        obj = JonesObject(J)

        # Remake the probe array
        probes = []
        for angle in pol_angles:
            rad = np.deg2rad(angle)
            jones_vec = torch.tensor([np.cos(rad) + 0j, np.sin(rad) + 0j], dtype=cdtype, device=device)
            probes.append(Probe(probe_amplitude, jones_vec, fluence=fluence, normalized=True))

        # Forward model
        model = ForwardModel(obj, Propagator(), Detector())
        I_pred = model.simulate_all(probes, scan)
        I_pred_safe = torch.clamp(I_pred, min=0.0)
        

        fluence_calc = torch.sum(torch.abs(probe_amplitude)**2)
        print(f"Calculated fluence from probe amplitude: {fluence_calc.item():.6e}, target fluence: {fluence.item():.6e}")
        '''
        if loss_prefactors['gradient_pf'] > 0 or loss_prefactors['anisotropy_pf'] > 0:
            #Only perform this computation if the prefactors are > 0, as it is quite expensive.

            # 2. Compute spatial gradients using finite differences
            # Differences along the X-axis (width)
            dx_lx = torch.diff(lx_norm, dim=-1)
            dx_ly = torch.diff(ly_norm, dim=-1)
            dx_lz = torch.diff(lz_norm, dim=-1)

            # Differences along the Y-axis (height)
            dy_lx = torch.diff(lx_norm, dim=-2)
            dy_ly = torch.diff(ly_norm, dim=-2)
            dy_lz = torch.diff(lz_norm, dim=-2)

            grad_mag_x = torch.sqrt(dx_lx**2 + dx_ly**2 + dx_lz**2 + 1e-8) # 1e-8 prevents NaN in sqrt at 0
            grad_mag_y = torch.sqrt(dy_lx**2 + dy_ly**2 + dy_lz**2 + 1e-8)

            loss_x = torch.sum(grad_mag_x**2)
            loss_y = torch.sum(grad_mag_y**2)

            gradient_loss = loss_x + loss_y
            anisotropy_loss = torch.sum(l[2]**2) # Positive prefactor will favour IP alignment.
        else:
            gradient_loss = 0.0
            anisotropy_loss = 0.0
        '''
        '''
        # Amplitude-based ptychographic loss
        if loss_prefactors['probe_localisation_pf'] > 0:
            probe_localisation_loss = loss_prefactors['probe_localisation_pf'] * torch.mean(torch.abs(probe_amplitude)**2*R**2) #This stops the probe from drifting by penalising amplitude at large R
        else:
            probe_localisation_loss = 0.0
        '''
        # Define the loss function
        loss = loss_prefactors['sqrt_amp_pf_c'] * torch.mean((torch.sqrt(I_pred_safe + eps) - torch.sqrt(I_meas + eps))**2) #+ loss_prefactors['gradient_pf'] * gradient_loss + loss_prefactors['anisotropy_pf'] * anisotropy_loss + probe_localisation_loss
        #loss = loss + loss_prefactors['STXM_pf']*(torch.sqrt(torch.sum(I_pred_safe + eps) - torch.sqrt(torch.sum(I_meas + eps)))**2) # This is a simple loss term to try to match the total intensity, which can help with convergence when the probe is also being optimised.
        # Loop over each diffraction pattern and measure the difference in the total intensity transmitted
        for pos in range(len(scan.positions)):
            for pol in range(len(pol_angles)):
                loss = loss+loss_prefactors['STXM_pf_c']*(torch.sqrt(torch.sum(I_pred_safe[pol, pos, :,:] + eps)) - torch.sqrt(torch.sum(I_meas[pol, pos, :,:] + eps)))**2
        
        loss = loss + loss_prefactors['probe_size_pf'] * torch.sum(torch.abs(probe_amplitude)*(R>R_probe))
        loss.backward()
        optimizer.step()

        # Step the scheduler with the current loss
        scheduler.step(loss)

        if iteration % 20 == 0:
            print([group['lr'] for group in optimizer.param_groups])
        
        if iteration % 8 == 0 and iteration < 30:
            with torch.no_grad():
                if probe_randomisation:
                    probe_amplitude.data += (torch.rand((H, W), device=device) * torch.max(torch.abs(probe_amplitude))*0.25)
                if object_randomisation:
                    for i in range(3):
                        noise = torch.randn_like(l[i])
                        l[i] += 1.5 * noise

        loss_history.append(loss.item())

        if iteration % 1 == 0:
            theta = torch.acos(torch.clamp(lz_norm, -1.0 + 1e-6, 1.0 - 1e-6)) # Clamp to avoid NaNs from acos outside of [-1, 1]
            phi = torch.atan2(ly_norm, lx_norm)
            l_mag = torch.sqrt(torch.sum(l**2, dim=0) + eps)
            print(f"Iter {iteration:4d} | Loss = {loss.item():.6e} | C, A1, A2: {C.item():.6e}, {A1.item():.6e}, {A2.item():.6e}")
            plot_update(probe_amplitude, theta, phi, loss, save_filename=plot_filename)

    return {
        "l": l.detach(),
        "probe_amplitude": probe_amplitude.detach(),
        "C": C.detach(),
        "A1": A1.detach(),
        "A2": A2.detach(),
        "loss_history": loss_history,
        "optimizer_params": optimizer_params,
        "loss_prefactors": loss_prefactors,   
    }
```

```python
import torch
import numpy as np


class PtychoReconstructionTrainer:
    def __init__(
        self,
        scan,
        pol_angles,
        I_meas,
        H,
        W,
        initial_l,
        initial_probe_amplitude,
        C,
        A1,
        A2,
        device=None,
        loss_prefactors=None,
        optimizer_params=None,
        Lx=None,
        Ly=None,
        R_probe=None,
        R=None,
        fluence=None,
        plot_update=None,
        phi_cmap=None,
        theta_cmap='magma',
        object_randomisation=False,
        probe_randomisation=False,
    ):
        """
        Initialize the ptychographic reconstruction trainer with full support for 
        object and probe optimization.
        
        Parameters
        ----------
        scan : ScanTrajectory
            Scan positions.
        pol_angles : list of float
            List of polarisation angles in degrees for each probe.
        I_meas : torch.Tensor
            Measured diffraction intensities, shape (N_probes, N_scan, Hdet, Wdet)
        H, W : int
            Object map size.
        initial_l : torch.Tensor
            Initial guess for the neel vector in Cartesian coordinates, shape (3, H, W).
        initial_probe_amplitude : torch.Tensor
            Initial guess for the probe amplitude, shape (H, W).
        C, A1, A2 : torch.Tensor or complex
            XMLD/Jones constants.
        device : torch.device or str
            Device to use.
        loss_prefactors : dict
            Prefactors for different loss components.
        optimizer_params : dict
            Learning rates for different parameter groups.
        Lx, Ly : float
            Object dimensions.
        R_probe : float
            Probe radius for regularization.
        R : torch.Tensor
            Radial distance grid.
        fluence : float or torch.Tensor
            Target fluence.
        plot_update : callable
            Function to call for live plot updates.
        phi_cmap : colormap
            Colormap for phi visualization.
        theta_cmap : str
            Colormap name for theta visualization.
        object_randomisation : bool
            Whether to apply randomisation to object.
        probe_randomisation : bool
            Whether to apply randomisation to probe.
        """
        self.scan = scan
        self.pol_angles = pol_angles
        self.I_meas = I_meas
        self.H = H
        self.W = W
        self.Lx = Lx
        self.Ly = Ly
        self.R_probe = R_probe
        self.R = R
        self.fluence = fluence
        self.plot_update = plot_update
        self.phi_cmap = phi_cmap
        self.theta_cmap = theta_cmap

        self.device = device or I_meas.device

        self.loss_prefactors = loss_prefactors or {
            'sqrt_amp_pf': 1.0, 
            'gradient_pf': 0.0,
            'anisotropy_pf': 0.0, 
            'probe_localisation_pf': 0.0,
            'STXM_pf': 1.0,
            'probe_size_pf': 1e-2
        }
        self.optimizer_params = optimizer_params or {
            'l_lr': 1e-3, 
            'CA_lr': 1e-5,
            'probe_lr': 1e-2
        }

        self.eps = 1e-6
        self.cdtype = torch.complex64

        # Parameters
        self.l = torch.nn.Parameter(initial_l.to(self.device))
        self.probe_amplitude = torch.nn.Parameter(initial_probe_amplitude.to(self.device))

        # Initialize C, A1, A2 as parameters
        if isinstance(C, torch.Tensor):
            self.C = torch.nn.Parameter(C.to(self.device))
        else:
            self.C = torch.nn.Parameter(torch.tensor(C, dtype=self.cdtype, device=self.device))
        
        if isinstance(A1, torch.Tensor):
            self.A1 = torch.nn.Parameter(A1.to(self.device))
        else:
            self.A1 = torch.nn.Parameter(torch.tensor(A1, dtype=self.cdtype, device=self.device))
        
        if isinstance(A2, torch.Tensor):
            self.A2 = torch.nn.Parameter(A2.to(self.device))
        else:
            self.A2 = torch.nn.Parameter(torch.tensor(A2, dtype=self.cdtype, device=self.device))

        self.object_randomisation = object_randomisation
        self.probe_randomisation = probe_randomisation

        self._init_optimizers()

        self.loss_history = []
        self.start_iter = 0

    def _init_optimizers(self):
        """Initialize optimizer and scheduler."""
        self.optimizer = torch.optim.AdamW([
            {"params": [self.l], "lr": self.optimizer_params['l_lr']},
            {"params": [self.probe_amplitude], "lr": self.optimizer_params['probe_lr']}
        ])

        self.scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
            self.optimizer, 
            mode='min',
            factor=0.5,
            patience=500,
            threshold=1e-3,
            min_lr=1e-6,
        )

    def save_checkpoint(self, path, iteration):
        """Save checkpoint with all model state and optimizer state."""
        checkpoint = {
            "iteration": iteration,
            "l": self.l.detach().cpu(),
            "probe_amplitude": self.probe_amplitude.detach().cpu(),
            "C": self.C.detach().cpu(),
            "A1": self.A1.detach().cpu(),
            "A2": self.A2.detach().cpu(),
            "optimizer_state_dict": self.optimizer.state_dict(),
            "scheduler_state_dict": self.scheduler.state_dict(),
            "loss_history": self.loss_history,
            "optimizer_params": self.optimizer_params,
            "loss_prefactors": self.loss_prefactors,
            "rng_state": torch.get_rng_state(),
            "cuda_rng_state": torch.cuda.get_rng_state_all() if torch.cuda.is_available() else None,
        }
        torch.save(checkpoint, path)

    def load_checkpoint(self, checkpoint):
        """Load checkpoint and restore all state."""
        self.l.data = checkpoint["l"].to(self.device)
        self.probe_amplitude.data = checkpoint["probe_amplitude"].to(self.device)
        self.C.data = checkpoint["C"].to(self.device)
        self.A1.data = checkpoint["A1"].to(self.device)
        self.A2.data = checkpoint["A2"].to(self.device)

        self.optimizer.load_state_dict(checkpoint["optimizer_state_dict"])
        self.scheduler.load_state_dict(checkpoint["scheduler_state_dict"])

        torch.set_rng_state(checkpoint["rng_state"])
        if torch.cuda.is_available() and checkpoint["cuda_rng_state"] is not None:
            torch.cuda.set_rng_state_all(checkpoint["cuda_rng_state"])

        self.loss_history = checkpoint.get("loss_history", [])
        self.start_iter = checkpoint["iteration"] + 1

    def _build_normalized_components(self):
        """Build normalized Cartesian components from l."""
        l_mag = torch.sqrt(torch.sum(self.l ** 2, dim=0) + 1e-8)
        lx_norm = self.l[0] / l_mag
        ly_norm = self.l[1] / l_mag
        lz_norm = self.l[2] / l_mag
        return lx_norm, ly_norm, lz_norm, l_mag

    def _build_probes(self, pol_angles):
        """Build probe array for all polarisation angles."""
        probes = []
        for angle in pol_angles:
            rad = np.deg2rad(angle)
            jones_vec = torch.tensor(
                [np.cos(rad) + 0j, np.sin(rad) + 0j], 
                dtype=self.cdtype, 
                device=self.device
            )
            probes.append(Probe(self.probe_amplitude, jones_vec, fluence=self.fluence, normalized=True))
        return probes

    def forward_model(self):
        """
        Compute forward model predictions.
        
        Returns
        -------
        torch.Tensor
            Predicted diffraction intensities, shape (N_probes, N_scan, Hdet, Wdet)
        """
        lx_norm, ly_norm, lz_norm, _ = self._build_normalized_components()

        # Build Jones object from current parameters
        neel = NeelObject(self.C, self.A1, self.A2)
        J = neel.build_jones_from_cartesian(lx=lx_norm, ly=ly_norm, lz=lz_norm)
        obj = JonesObject(J)

        # Build probes
        probes = self._build_probes(self.pol_angles)

        # Forward model
        model = ForwardModel(obj, Propagator(), Detector())
        I_pred = model.simulate_all(probes, self.scan)
        
        return I_pred

    def compute_loss(self, I_pred, iteration=0):
        """
        Compute total loss with multiple components.
        
        Parameters
        ----------
        I_pred : torch.Tensor
            Predicted diffraction intensities.
        iteration : int
            Current iteration (used for scheduling loss prefactors).
            
        Returns
        -------
        torch.Tensor
            Total loss value.
        """
        I_pred_safe = torch.clamp(I_pred, min=0.0)

        # Get current loss prefactors (may vary with iteration)
        sqrt_amp_pf = self.loss_prefactors.get('sqrt_amp_pf_c', self.loss_prefactors['sqrt_amp_pf'])
        STXM_pf = self.loss_prefactors.get('STXM_pf_c', self.loss_prefactors['STXM_pf'])

        # Amplitude-based ptychographic loss
        loss = sqrt_amp_pf * torch.mean(
            (torch.sqrt(I_pred_safe + self.eps) - torch.sqrt(self.I_meas + self.eps)) ** 2
        )

        # STXM loss: match total intensity transmission at each position for each probe
        for pos in range(len(self.scan.positions)):
            for pol in range(len(self.pol_angles)):
                loss = loss + STXM_pf * (
                    torch.sqrt(torch.sum(I_pred_safe[pol, pos, :, :] + self.eps)) - 
                    torch.sqrt(torch.sum(self.I_meas[pol, pos, :, :] + self.eps))
                ) ** 2

        # Probe localization loss: keep probe amplitude localized near center
        if self.loss_prefactors.get('probe_size_pf', 0) > 0 and self.R is not None:
            loss = loss + self.loss_prefactors['probe_size_pf'] * torch.sum(
                torch.abs(self.probe_amplitude) * (self.R > self.R_probe)
            )

        return loss

    def train(self, num_iterations, checkpoint_out_path=None, save_every=50, plot_filename=None):
        """
        Full training loop with complex scheduling and optional randomization.
        
        Parameters
        ----------
        num_iterations : int
            Number of iterations to train.
        checkpoint_out_path : str or None
            Path to save checkpoints.
        save_every : int
            Save checkpoint every N iterations.
        plot_filename : str or None
            Filename for saving plots during training.
        """
        STXM_end_iteration = 50

        for iteration in range(self.start_iter, num_iterations):
            # Conditionally freeze/unfreeze parameters based on iteration
            if iteration < STXM_end_iteration:
                # Free object parameters, fix probe
                for p in [self.l, self.C, self.A1, self.A2]:
                    p.requires_grad_(True)
                self.probe_amplitude.requires_grad_(False)
                self.loss_prefactors['STXM_pf_c'] = self.loss_prefactors['STXM_pf']
                self.loss_prefactors['sqrt_amp_pf_c'] = self.loss_prefactors['sqrt_amp_pf']
                if iteration == 0:
                    print('Free object parameters, fixing probe amplitude.')
            elif iteration >= 50 and iteration % 2 == 0:
                # Alternate: update object on even iterations
                for p in [self.l, self.C, self.A1, self.A2]:
                    p.requires_grad_(True)
                self.probe_amplitude.requires_grad_(False)
            elif iteration >= 50 and iteration % 2 == 1:
                # Alternate: update probe on odd iterations
                for p in [self.l, self.C, self.A1, self.A2]:
                    p.requires_grad_(False)
                self.probe_amplitude.requires_grad_(True)
            
            if iteration >= 80:
                # Free all parameters at the end
                for p in [self.l, self.C, self.A1, self.A2]:
                    p.requires_grad_(True)
                self.probe_amplitude.requires_grad_(True)
                if iteration == 80:
                    print('Freeing all parameters.')

            self.optimizer.zero_grad(set_to_none=True)

            # Forward pass
            I_pred = self.forward_model()
            loss = self.compute_loss(I_pred, iteration)

            # Backward pass
            loss.backward()
            self.optimizer.step()
            self.scheduler.step(loss)

            self.loss_history.append(loss.item())

            # Print learning rates periodically
            if iteration % 20 == 0:
                print([group['lr'] for group in self.optimizer.param_groups])

            # Apply randomization early in training to escape local minima
            if iteration % 8 == 0 and iteration < 30:
                with torch.no_grad():
                    if self.probe_randomisation:
                        self.probe_amplitude.data += (
                            torch.rand((self.H, self.W), device=self.device) * 
                            torch.max(torch.abs(self.probe_amplitude)) * 0.25
                        )
                    if self.object_randomisation:
                        for i in range(3):
                            noise = torch.randn_like(self.l[i])
                            self.l[i].data += 1.5 * noise

            # Save checkpoint
            if checkpoint_out_path and iteration % save_every == 0:
                self.save_checkpoint(checkpoint_out_path, iteration)

            # Update plots
            if iteration % 1 == 0 and self.plot_update is not None:
                lx_norm, ly_norm, lz_norm, l_mag = self._build_normalized_components()
                theta = torch.acos(torch.clamp(lz_norm, -1.0 + 1e-6, 1.0 - 1e-6))
                phi = torch.atan2(ly_norm, lx_norm)
                
                # Compute fluence from probe
                fluence_calc = torch.sum(torch.abs(self.probe_amplitude) ** 2)
                print(
                    f"Iter {iteration:4d} | Loss = {loss.item():.6e} | "
                    f"C, A1, A2: {self.C.item():.6e}, {self.A1.item():.6e}, {self.A2.item():.6e} | "
                    f"Fluence: {fluence_calc.item():.6e}"
                )
                
                self.plot_update(self.probe_amplitude, theta, phi, loss, save_filename=plot_filename)

    def get_results(self):
        """Get reconstructed parameters and loss history."""
        return {
            "l": self.l.detach(),
            "probe_amplitude": self.probe_amplitude.detach(),
            "C": self.C.detach(),
            "A1": self.A1.detach(),
            "A2": self.A2.detach(),
            "loss_history": self.loss_history,
            "optimizer_params": self.optimizer_params,
            "loss_prefactors": self.loss_prefactors,
        }
```
